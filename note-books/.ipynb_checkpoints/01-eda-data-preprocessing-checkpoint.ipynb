```python
### Essential libraries we use
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OrdinalEncoder
from sklearn.impute import KNNImputer
import pickle
%matplotlib inline
```

```python
def death_rate(column, category):
    dead = data[(data[column] == category) & (data["stroke"] == 1)].shape[0]
    total = data[(data[column] == category)].shape[0]
    result = dead / total
    return result * 100
```

```python
# Load data
df = pd.read_csv("../data/healthcare-dataset-stroke-data.csv", index_col = 0)
```

```python
df.head()
```

```python
# Check the columns adn their missing values
df.info()
```

```python
# Two types of features we have here
categorical = ["gender", "hypertension", "heart_disease", "ever_married", "work_type", "Residence_type",
              "smoking_status", "stroke"]
continuous = ["age", "avg_glucose_level", "bmi"]
```

```python
# Check descriptive statistics
df.describe()
```

```python
# Values of categorical features
for col in categorical:
    print(df[col].value_counts())
    print("--------------------------")
```

```python
# Remove the row with undefined gender
df.drop(df[df["gender"] == "Other"].index, inplace=True)
```

```python
# Change Unknown  to Nan for imputation
df["smoking_status"].replace("Unknown", np.nan, inplace=True)
```

```python
data = df.copy()
```

```python
cols = categorical[:-1]
```

```python
# Encoding categorical values for further analysis and modeling
encoder = OrdinalEncoder(handle_unknown = "use_encoded_value", unknown_value=np.nan)
```

```python
encoder.fit(df[cols])
```

```python
data[cols] = encoder.transform(df[cols])
```

```python
data.head()
```

```python
data = data.iloc[:,:-1].copy()
```

```python
data.head()
```

```python
imputer = KNNImputer(n_neighbors = 2)
```

```python
imputer.fit(data)
```

```python
data = pd.DataFrame(imputer.transform(data), columns = data.columns, index = data.index)
```

```python
data = data.join(df.iloc[:,-1])
```

```python
data.head()
```

```python
data.info()
```

```python
# Set categorical variables to int32 for a bit of memory efficiency
data[categorical] = data[categorical].astype(np.int32)
```

```python
data.info()
```

```python
pd.plotting.scatter_matrix(data[continuous], figsize=(8,8))
plt.plot()
```

```python
sns.heatmap(data.corr())
```

```python
stroke_0 = data.loc[df["stroke"] == 0]
stroke_1 = data.loc[df["stroke"] == 1]
plt.scatter(stroke_0.loc[:,"age"], stroke_0.loc[:, "bmi"], c="blue", alpha=.5)
plt.scatter(stroke_1.loc[:,"age"], stroke_1.loc[:, "bmi"], c="red", alpha = .5)
plt.plot()
```

```python
plt.scatter(stroke_0.loc[:,"age"], stroke_0.loc[:, "avg_glucose_level"], c="blue", alpha=.5)
plt.scatter(stroke_1.loc[:,"age"], stroke_1.loc[:, "avg_glucose_level"], c="red", alpha = .5)
plt.plot()
```

```python
data[cols] = encoder.inverse_transform(data[cols])
```

```python
data.head()
```

```python
data.value_counts
```

```python
# Values of categorical features
for col in categorical:
    print(data[col].value_counts())
    print("--------------------------")
```

```python
fig, axs = plt.subplots(3,2, figsize=(15, 13))
counter = 0
for i in range(3):
    for j in range(2):
        column = categorical[counter]
        categories = data[column].unique().tolist()
        rates = [death_rate(column, category) for category in categories]
        axs[i,j].set_title(column)
        axs[i,j].bar(categories, rates, width=0.3)
        counter += 1
plt.subplots_adjust(hspace=0.5, wspace=0.4)  # Adjust the hspace and wspace values as desired
fig.suptitle("Death rate for different categories of each Column")
plt.plot()
```

```python
data[cols] = encoder.fit_transform(data[cols])
```

```python
data[categorical] = data[categorical].astype(np.int32)
```

```python
data.head()
```

```python
data.to_csv("../data/stroke-data-preprocessed.csv")
```
